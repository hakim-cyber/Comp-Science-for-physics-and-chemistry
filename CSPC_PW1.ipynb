```python
import numpy as np
import matplotlib.pyplot as plt

#trajectory

t = np.linspace(0, 20, 500)
dt = t[1] - t[0]


x = 50 * np.sin(0.1 * np.pi * t)
y = 50 * np.sin(0.2 * np.pi * t)


plt.figure(figsize=(4,4))
plt.plot(x, y)
plt.show()

# accelerations(we take second derivative)

ax = -0.5 * np.pi**2 * np.sin(0.1 * np.pi * t)
ay = -2.0 * np.pi**2 * np.sin(0.2 * np.pi * t)



plt.plot(t, ax)
plt.plot(t, ay)
plt.show()
```

Exercise 2


```python
noise_x = np.random.normal(5,0.001, len(x))
noise_y = np.random.normal(5,0.001, len(y))

x_noisy = x + noise_x
y_noisy = y + noise_y

plt.figure(figsize=(4,4))
plt.plot(x_noisy, y_noisy)
plt.show()


#velocity numerical differentiation
vx_noise = np.gradient(x_noisy, dt)
vy_noise = np.gradient(y_noisy, dt)

#acceleration numerical differentiation
ax_noise = np.gradient(vx_noise, dt)
ay_noise = np.gradient(vy_noise, dt)

#comparing ax and noisy ax
plt.figure(figsize=(4, 4))
plt.plot(t, ax, color='black', linewidth=2)
plt.plot(t, ax_noise, 'o', markersize=3, color='blue')
plt.show()

#comparing ay and noisy ay
plt.figure(figsize=(4, 4))
plt.plot(t, ay, color='black', linewidth=2)
plt.plot(t, ay_noise, 'o', markersize=3, color='blue')
plt.show()


# assesing impact of offset and noise on accelerations

diff_ax = ax - ax_noise
diff_ay = ay - ay_noise

plt.figure(figsize=(4, 4))
plt.plot(t, diff_ax, color='red')
plt.plot(t, diff_ay, color='blue')
plt.show()
```

exercise 3


```python
ax_noise2 = ax + np.random.normal(0.1, 1, len(t))
ay_noise2 = ay + np.random.normal(0.1, 1, len(t))

#for simplictiy we assume ideal
vx0 = (x[1] - x[0]) / dt
vy0 = (y[1] - y[0]) / dt
x0  = x[0]
y0  = y[0]

# accelaration- speed
vx_noise2 = vx0 + np.cumsum(ax_noise2) * dt
vy_noise2 = vy0 + np.cumsum(ay_noise2) * dt

# speed- position
x_noise2 = x0 + np.cumsum(vx_noise2) * dt
y_noise2 = y0 + np.cumsum(vy_noise2) * dt

#comparing x  and x noise2
plt.figure(figsize=(4, 4))
plt.plot(t, x, color='black', linewidth=2)
plt.plot(t,x_noise2, 'o', markersize=3, color='blue')
plt.show()

#comparing y  and y noise2
plt.figure(figsize=(4, 4))
plt.plot(t, y, color='black', linewidth=2)
plt.plot(t,y_noise2, 'o', markersize=3, color='blue')
plt.show()

# assesing impact of offset and noise on position

diff_x_2 = x - x_noise2
diff_y_2 = y - y_noise2

plt.figure(figsize=(4, 4))
plt.plot(t, diff_x_2, color='red')
plt.plot(t, diff_y_2, color='blue')
plt.show()
```
