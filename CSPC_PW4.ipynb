```python
import sympy as sp

# consts
VA, VB, VC, VD = sp.symbols('VA VB VC VD', real=True)
VCC, R1, R2, R3, R4 = sp.symbols('VCC R1 R2 R3 R4', positive=True)

#equations
eq1 = sp.Eq(VA, (VCC/R1  + VC/R3 + VB/R3 + VD/R4) / (1/R1 + 2/R3 + 1/R4))
eq2 = sp.Eq(VB, (VCC/R2  + VD/R3 + VA/R3) / (1/R2 + 2/R3 ))
eq3 = sp.Eq(VC, (VA/R3 + VD/R3) / (1/R2 + 2/R3) )
eq4 = sp.Eq(VD, (VB/R3 +  VC/R3 + VA/R4) / ( 1/R1 + 2/R3  + 1/R4) )

#solve
sol = sp.solve([eq1, eq2, eq3, eq4], [VA, VB, VC,  VD], dict=True )[0]

# values
subs = {VCC: 15, R1: 1000, R2: 2000, R3:  10000, R4: 500}
numeric_solution = {k: float(v.subs(subs)) for  k, v in sol.items()}

for name, value in numeric_solution.items():
    print(f"{name} = {value:.6f} V")
```

```python

import numpy as np

# consts
VCC = 15
R1, R2, R3, R4 = 1000, 2000, 10000, 500

# conductance
g1, g2, g3, g4 = 1/R1, 1/R2, 1/R3, 1/R4

A = np.array([
    [g1+2*g3+g4, -g3,        -g3,        -g4],
    [-g3,        g2+2*g3,    0,          -g3],
    [-g3,        0,          g2+2*g3,    -g3],
    [-g4,        -g3,        -g3,        g1+2*g3+g4]
], dtype=float)

B = np.array([g1*VCC, g2*VCC, 0, 0], dtype=float)

x = np.linalg.solve(A, B)

labels = ["VA", "VB", "VC", "VD"]

for lbl, val in zip(labels, x):
    print(f"{lbl} = {val:.6f} V")
```

```python
import numpy as np
from scipy.optimize import minimize

#const
VCC = 15
R1, R2, R3, R4 = 1000, 2000, 10000, 500

def residuals(x):
    VA, VB, VC, VD = x
    g1, g2, g3, g4 = 1/R1, 1/R2, 1/R3, 1/R4
    denom1 = g1 + 2*g3 + g4
    denom2 = g2 + 2*g3
    r1 = VA - (VCC*g1 + VC*g3 + VB*g3 + VD*g4) / denom1
    r2 = VB - (VCC*g2 + VD*g3 + VA*g3) / denom2
    r3 = VC - (VA*g3 + VD*g3) / denom2
    r4 = VD - (VB*g3 + VC*g3 + VA*g4) / denom1
    return np.array([r1, r2, r3, r4])

def cost(x):
    r = residuals(x)
    return np.dot(r, r)


x0 = np.array([VCC/2]*4)
result = minimize(cost, x0, method='BFGS')


VA, VB, VC, VD = result.x

print(f"VA = {VA:.6f} V")
print(f"VB = {VB:.6f} V")
print(f"VC = {VC:.6f} V")
print(f"VD = {VD:.6f} V")
```
